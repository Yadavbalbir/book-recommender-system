```python
import numpy as np
import pandas as pd
```

```python
books = pd.read_csv('BX-Books.csv', sep=";", error_bad_lines = False, encoding= 'latin-1')
```

```python
books.head()
```

```python
books=books[['ISBN', 'Book-Title', 'Book-Author','Year-Of-Publication','Publisher']]
```

```python
books.head()
```

```python
books.rename(columns={
    'Book-Title':'title',
    'Book-Author':'author',
    'Year-Of-Publication': 'year',
    'Publisher':'publisher'
}, inplace=True)
books.head()
```

```python
users=pd.read_csv('BX-Users.csv', sep=";", error_bad_lines = False, encoding= 'latin-1')
```

```python
users.head()
```

```python
users.rename(columns={
    'User-ID': 'user_id',
    'Location': 'location',
    'Age': 'age',
}, inplace=True)

users.head()
```

```python
ratings = pd.read_csv('BX-Book-Ratings.csv',  sep=";", error_bad_lines = False, encoding= 'latin-1')
```

```python
ratings.head()
```

```python
ratings.rename(columns={
    'User-ID':'user_id',
    'Book-Rating':'rating'
}, inplace=True)
```

```python
ratings.head()
```

```python
books.shape
```

```python
users.shape
```

```python
ratings.shape
```

```python
selected_user=ratings['user_id'].value_counts()>250
```

```python
selected_user[selected_user].shape
```

```python
selected_user_id=selected_user[selected_user].index
selected_user_id
```

```python
ratings=ratings[ratings['user_id'].isin(selected_user_id)]
```

```python
ratings.shape
```

```python
rating_with_books = ratings.merge(books, on='ISBN')
```

```python
rating_with_books
```

```python
number_of_rating = rating_with_books.groupby('title')['rating'].count().reset_index().rename(columns={
    'rating' : 'number_of_rating',
})
```

```python
number_of_rating.head()
```

```python
final_rating = rating_with_books.merge(number_of_rating, on="title")
```

```python
final_rating = final_rating[final_rating['number_of_rating']>=50]
```

```python
final_rating
```

```python
final_rating.drop_duplicates(['user_id','title'], inplace=True)
```

```python
final_rating
```

```python
book_pivot = final_rating.pivot_table(columns='user_id', index='title', values='rating')
```

```python
book_pivot
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from scipy.sparse import csr_matrix
book_sparse = csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm= 'brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(book_pivot.iloc[157,:].values.reshape(1,-1), n_neighbors=6)
```

```python
    distances
```

```python
suggestions
```

```python
result=[]
for i in range(len(suggestions)):
    result.append(book_pivot.index[suggestions[i]])
```

```python
result
```

```python
for i in range(len(result[0])):
    print(result[0][i])
```

```python
def recommend(title):
    user_id=np.where(book_pivot.index==title)[0][0]
    distances, suggestions = model.kneighbors(book_pivot.iloc[user_id,:].values.reshape(1,-1), n_neighbors=6)
    result=[]
    for i in range(len(suggestions)):
        result.append(book_pivot.index[suggestions[i]])
    for i in range(1,len(result[0])):
        print(result[0][i])
    
```

# Search Recommended Book
    recommend('type book title')
